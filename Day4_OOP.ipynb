#### Class Methods

```python
class Employee:
    salary_raise = 1.10
    
    def __init__(self, first, last, salary):
        self.first = first
        self.last = last
        self.salary = salary
        self.email = first + "." + last + "@company.org"
        
    # Instance Method
    def full_name(self):
        return f'{self.first} {self.last}'
    
    # Method to increment an instance attribute's value
    def apply_raise(self):
        self.salary = int(self.salary * self.salary_raise)
    
    # Method to change an instance attribute's value
    def change_last_name(self, new_last):
        self.last = new_last
        self.email = self.first + "." + new_last + "@company.org"
        
        
emp_1 = Employee('Justin', 'Fields', 100000)
emp_2 = Employee('Andy', 'Dalton', 200000)

employees = [emp_1, emp_2]

for e in employees:
    print(e.full_name())
    print(e.email)
```

```python
for emp in employees:
    print('Current salary:')
    print(emp.salary)
    print('Applying raise...')
    emp.apply_raise()
    print('New Salary:')
    print(emp.salary)
    print('=========================')
```

```python
# Add two methods to the Car class. The first method will add miles to the car based on how far it drove. The second will
# print out all of the info for the car in a formatted way e.g. '2021 Honda Accord with 150 miles'


class Car:
    wheels = 4 # Class Attribute
    
    def __init__(self, make, model, year, drivers=[], miles=0):
        self.make = make # Instance Attribute
        self.model = model # Instance Attribute
        self.year = year # Instance Attribute
        self.miles = miles
        self.drivers = drivers
        
    def __str__(self):
        return f"Car object: {self.year} {self.make} {self.model}"
    
    def __repr__(self):
        return f"<Car | {self.make} {self.model}>"
        
    def add_miles(self, miles_driven):
        self.miles = self.miles + miles_driven
    
    def info(self):
        return f'{self.year} {self.make} {self.model} with {self.miles} miles.'
        
car1 = Car('Honda', 'Accord', 2021, drivers=['Brian', 'Dan', 'Fred'])
car2 = Car('Toyota', 'Prius', 2020)
car3 = Car('Dodge', 'Viper', 2017)

my_cars = [car1, car2, car3]
```

### Exercise 1 - Create an Object Oriented Shopping Cart

```python
# Create a class called cart that retains items and has methods to add, remove, and show

class Cart():
    def __init__(self):
        self.cart = []
        
    def add_to_cart(self, item):
        self.cart.append(item)
        
    def remove_from_cart(self, item):
        if item not in self.cart:
            print(f"{item} is not in cart")
        else:
            self.cart.remove(item)
            
    def show_cart(self):
        for item in self.cart:
            print(item)
            
    def clear_cart(self):
        self.cart.clear()

        

def run():
    my_cart = Cart()
    while True:
        response = input('What would you like to do? Add/Remove/Show/Clear or Quit? ').lower()
        if response == 'quit':
            print('Thank you for shopping with us. Here is what you purchased:')
            my_cart.show_cart()
            break
        elif response == 'add':
            item_to_add = input('What item would you like to add to your cart? ')
            my_cart.add_to_cart(item_to_add)
            print(my_cart.cart)
        elif response == 'remove':
            item_to_remove = input('What item would you like to remove from your cart? ')
            my_cart.remove_from_cart(item_to_remove)
            print(my_cart.cart)
        elif response == 'show':
            my_cart.show_cart()
        elif response == 'clear':
            my_cart.clear_cart()
            print('Your cart has been cleared.')
        else:
            print(f'{response} is not a valid option, please try again.')
            

run()
```

```python
# Create a Parent Class

class Animal:
    kingdom = 'Animal'
    
    def __init__(self, name, species, legs=4):
        print('Animal __init__ called')
        self.name = name
        self.species = species
        self.legs = legs
        
    def make_sound(self, sound):
        print(f'{self.name} says {sound}')
        

# Create child classes
class Dog(Animal):
    def __init__(self, name, species, color, legs=4):
        print('Dog __init__ called')
        super().__init__(name, species)
        self.color = color
        
    def fetch(self):
        print(f'{self.name} has fetched')
        
        
dog = Dog('Buddy', 'dog', 'golden')
lion = Animal('Simba', 'lion')
```

```python
class Cart():
    def __init__(self):
        self.cart = []
        
    def add_to_cart(self, item):
        self.cart.append(item)
        
    def remove_from_cart(self, item):
        if item not in self.cart:
            print(f"{item} is not in cart")
        else:
            self.cart.remove(item)
            
    def show_cart(self):
        for item in self.cart:
            print(item)
            
    def clear_cart(self):
        self.cart.clear()

class Item():
    def __init__(self, name, price):
        self.name = name
        self.price = price
        
    def __str__(self):
        return f'{self.name} - ${self.price}'
        

def run():
    my_cart = Cart()
    while True:
        response = input('What would you like to do? Add/Remove/Show/Clear or Quit? ').lower()
        if response == 'quit':
            print('Thank you for shopping with us. Here is what you purchased:')
            my_cart.show_cart()
            break
        elif response == 'add':
            name_of_item = input('What item would you like to add to your cart? ')
            price_of_item = input('How much does it cost? ')
            item_to_add = Item(name_of_item, price_of_item)
            my_cart.add_to_cart(item_to_add)
            print(my_cart.cart)
        elif response == 'remove':
            item_to_remove = input('What item would you like to remove from your cart? ')
            my_cart.remove_from_cart(item_to_remove)
            print(my_cart.cart)
        elif response == 'show':
            my_cart.show_cart()
        elif response == 'clear':
            my_cart.clear_cart()
            print('Your cart has been cleared.')
        else:
            print(f'{response} is not a valid option, please try again.')
            

run()
```

```python
class Player:
    def __init__(self, name, height, weight):
        self.name = name
        self.height = height
        self.weight = weight
        
    def shoot(self):
        print(f'{self.name} shoots the ball')
        
    def __repr__(self):
        return f'<Player|{self.name}>'
        
        
class Team:
    
    def __init__(self, pg, sg, sf, pf, c):
        self.pg = pg
        self.sg = sg
        self.sf = sf
        self.pf = pf
        self.c = c
        
    
russ = Player('Russel Westbrook', '6-1', '190')
steph = Player('Steph Curry', '6-1', '180')
kawhi = Player('Kawhi Leonard', '6-4', '200')
bobby = Player('Bobby Portis', '6-8', '210')
deandre = Player('Deandre Ayton', '6-10', '220')

all_stars = Team(russ, steph, kawhi, bobby, deandre)
```

```python
all_stars.c.shoot()
```

```python
all_stars.sg.shoot()
```
